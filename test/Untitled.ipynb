```python
import numpy as np
def compute_loss(x): 
    return  np.sum(x*x)
def compute_gradient(n,xt,xx,y,m):
    return n*np.dot(xt,(xx-y))/m

A=np.genfromtxt('data.txt',delimiter=',')
length = len(A)
#prepare data

x=np.array([np.ones(length),A[:,0]])
x=x.transpose()  
y =  np.array([A[:,1]])
y=y.transpose()

#process
theta=np.array([[0],[0]])
n=1e-17
for i in range(1,10000):
    xx=np.dot(x,theta)
    offset=y-xx
    loss_function=compute_loss(offset)/(2*length)
   # print(loss_function)
    xt=x.transpose()
    gradient=compute_gradient(n,xt,xx,y,length)
    theta=theta-gradient

print("theta ",theta)
print("loss function", loss_function)

#test data
print("so sanh du lieu")
kq=np.dot(x,theta)
for i in range(1,length):
    print(A[i][1]," ",kq[i][0]) 

    
    
    

```
